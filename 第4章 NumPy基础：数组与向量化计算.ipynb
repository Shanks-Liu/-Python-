__随机漫步__

```python
nwalks = 5000
nsteps = 1000
draws = np.random.randint(0, 2, size=(nwalks, nsteps))
steps = np.where(draws>0, 1, -1)
walks = steps.cumsum(1)
```

```python
walks
```

```python
walks.shape
```

```python
hists30 = (np.abs(walks) >= 30).any(1)
```

```python
hists30
```

```python
hists30.shape
```

```python
hists30.sum()
```

```python
crossing_times = (np.abs(walks[hists30]) >= 30).argmax(1)
```

```python
crossing_times.shape
```
